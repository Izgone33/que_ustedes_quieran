```python
#Importar librerias
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
bgr = cv2.imread('imagen.jpg')
print = ("El tamano de la imagen es: ", bgr.shape)
print

#Convertir a RGB
rgb = cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB)

#Mostrar imagen
plt.imshow(rgb)
```

```python
# Mostrar la imagen en escala de grises
gray = cv2.cvtColor(bgr, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
```

```python
#Mostrar los compornentes de color de la imagen
plt.figure(figsize=(10,5))
plt.subplot(1,3,1)
plt.imshow(bgr[:,:,0], cmap='gray')
plt.title('Canal Azul')
plt.subplot(1,3,2)
plt.imshow(bgr[:,:,1], cmap='gray')
plt.title('Canal Verde')
plt.subplot(1,3,3)
plt.imshow(bgr[:,:,2], cmap='gray')
plt.title('Canal Rojo')
```

```python
#Ajuste de brillo y contraste
alpha = 1.5 # Contraste
beta = 50   # Brillo
adjusted = cv2.convertScaleAbs(bgr, alpha=alpha, beta=beta)
plt.imshow(cv2.cvtColor(adjusted, cv2.COLOR_BGR2RGB))
plt.title('Imagen con brillo y contraste ajustados')
plt.show()
```

```python
#Aplicar filtro Gaussiano
blurred = cv2.GaussianBlur(bgr, (15, 15), 0)
plt.imshow(cv2.cvtColor(blurred, cv2.COLOR_BGR2RGB))
plt.title('Imagen con filtro Gaussiano')
plt.show()
```

```python
gray_bulur = cv2.GaussianBlur(gray, (15, 15), 0)
edges = cv2.Canny(gray_bulur, 50, 150)
plt.imshow(edges, cmap='gray')
plt.title('Deteccion de bordes con Canny')
plt.show()
```

```python
#deteccion de puntos de interes
sift = cv2.SIFT_create()
keypoints, descriptors = sift.detectAndCompute(gray, None)
img_keypoints = cv2.drawKeypoints(bgr, keypoints, None, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
plt.imshow(cv2.cvtColor(img_keypoints, cv2.COLOR_BGR2RGB))
plt.title('Puntos de interes detectados con SIFT')
plt.show()
```
